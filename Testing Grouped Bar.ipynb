```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = [25, 50, 75, 100, 125]
true_positive = [20, 34, 30, 35, 27]
false_positive = [25, 32, 34, 20, 25]
true_negative = [15, 12, 14, 10, 15] 
false_negative = [5, 2, 4, 1, 5]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars



fig, ax = plt.subplots(figsize=(10,5))

rects1 = ax.bar(x - width/4, true_positive, width, label='True Positive')
rects2 = ax.bar(x + width/4, false_positive, width, label='False Positive')
rects3 = ax.bar(x - width/4, true_negative, width, label='True Negative')
rects4 = ax.bar(x + width/4, false_negative, width, label='False Negative')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()
plt.show()
```
